```python
import math
import random
```

```python
def play():
    user_input = input("Enter your choice of Input: 'r' for Rock, 'p' for paper, and 's' for scissor\n")
    user = user_input.lower()
    
    input_char = ['r','p','s']
    computer = random.choice(input_char)
    
    result = is_win(user, computer)
    
def is_win(user, computer):
    player1_choice = user
    player2_choice = computer
    
    # winning conditions:
    winning_conditions = {
        ('r', 's'): True,
        ('s', 'p'): True,
        ('p', 'r'): True
    }
    
    if(player1_choice, player2_choice) in winning_conditions:
        print(" You chose {}. The computer chose {}. You won!".format(player1_choice, player2_choice))
        return True
    elif player1_choice == player2_choice:
        print(" You chose {}. The computer chose {}. It's a tie!".format(player1_choice, player2_choice))
        return None
    else:
        print("You chose {}. The computer chose {}. You lost!".format(player1_choice, player2_choice))
        return False
        
```

```python
play()
```
